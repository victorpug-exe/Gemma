```python
from transformers import pipeline
from PIL import Image
import requests
import torch

pipe = pipeline(
    "image-text-to-text",
    model="google/medgemma-4b-it",
    torch_dtype=torch.bfloat16,
    device="cuda",
)
```

```python
# Image attribution: Stillwaterising, CC0, via Wikimedia Commons
image_url = "https://upload.wikimedia.org/wikipedia/commons/c/c8/Chest_Xray_PA_3-8-2010.png"
image = Image.open(requests.get(image_url, headers={"User-Agent": "example"}, stream=True).raw)
```

```python
image
```

```python
messages = [
    {
        "role": "system",
        "content": [{"type": "text", "text": "Voc√™ √© um radiologista expert."}]
    },
    {
        "role": "user",
        "content": [
            {"type": "text", "text": "Descreva esse raio-x"},
            {"type": "image", "image": image}
        ]
    }
]

output = pipe(text=messages)
print(output[0]["generated_text"][-1]["content"])
```

```python
messages = [
    {
        "role": "system",
        "content": [{"type": "text", "text": "Voc√™ √© uma assistente m√©dica digital."}]
    },
    {
        "role": "user",
        "content": [
            {"type": "text", "text": "Como diferenciar pneumonia bacteriana de pneumonia viral?"},
        ]
    }
]

output = pipe(text=messages, max_new_tokens=1000)
print(output[0]["generated_text"][-1]["content"])
```
